```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use('ggplot')
```

```python
cric_data = pd.read_csv("matches.csv")
cric_data
```

```python
cols_to_drop = ['dl_applied']
cric_data.drop(cols_to_drop,axis=1,inplace=True)
cric_data
```

```python
print('No. of matches played:',cric_data.shape[0])
```

```python
print('No. of Seasons:',len(cric_data.season.unique()))
```

## No. of Matches played in each Season

```python
plt.figure(figsize=(15,5))
sns.set_theme(style='dark')
sns.countplot(x='season',saturation=0.95,data=cric_data)
plt.xticks(fontname='Times new Roman',fontsize=15)
plt.yticks(fontsize=15,fontname='Times New Roman')
plt.xlabel('Season',fontname='Times New Roman',fontsize=22)
plt.ylabel('Count',fontname='Times New Roman',fontsize=22)
plt.title('Matches Played in Each Season ',fontname='Times New Roman',fontsize=42)
plt.show()
```

### from graph we see that most no of matches played in 2013 followed by 2012 & 2011

## No. of Matches in each City

```python
plt.figure(figsize=(15,5))
sns.countplot(x='city',data=cric_data)
plt.xticks(rotation=90,fontname='Times new Roman',fontsize=12)
plt.yticks(fontsize=12,fontname='Times New Roman')
plt.xlabel('City',fontname='Times New Roman',fontsize=22)
plt.ylabel('Count',fontname='Times New Roman',fontsize=22)
plt.title('Matches Played in Each City ',fontname='Times New Roman',fontsize=42)
plt.show()
```

### In Mumbai most no of matches played of ipl

## No. of Matches at each Venue

```python
plt.figure(figsize=(15,5))
sns.countplot(x='venue',data=cric_data)
plt.xticks(rotation=90,fontsize=14,fontname='Times New Roman')
plt.yticks(fontsize=14,fontname='Times New Roman')
plt.xlabel('Venue',fontname='Times New Roman',fontsize=22)
plt.ylabel('Count',fontname='Times New Roman',fontsize=22)
plt.title('Matches Played at Each Venue ',fontname='Times New Roman',fontsize=42)
plt.show()
```

### Most no of matches played in M Chinnaswamy followed by Eden Gardens & Feroj Shah Kotla

## No. of matches played by each team

```python
team_df= pd.melt(cric_data,id_vars=['id','season'],value_vars=['team1','team2'])
team_df.rename(columns={'value':'teams'},inplace=True)
team_df
```

```python
team_data = team_df.teams.value_counts()
team_data
```

```python
team_data.plot(kind='bar',
            figsize=(16,5),
            color='teal')
plt.xticks(rotation=90,fontsize=14,fontname='Times New Roman')
plt.yticks(fontsize=14,fontname='Times New Roman')
plt.xlabel('Teams',fontname='Times New Roman',fontsize=24)
plt.ylabel('Count',fontname='Times New Roman',fontsize=24)
plt.title('Matches Played by Each Team ',fontname='Times New Roman',fontsize=42)
ax=plt.axes()
ax.set_facecolor('gray')
plt.show()
```

### Mumbai Indians played most no of matches followed by RCB & KKR

# Toss Decision

```python
toss_data = cric_data.toss_decision.value_counts()
toss_data.plot(kind='pie',
               labels=('Field','Bat'),
               figsize=(16,7),
               autopct='%.1f%%',
               colors=('DarkOliveGreen','Yellow'),
               startangle=90,
               shadow=True,
               textprops={'fontsize':30,'fontname':'Times New Roman'})
plt.axis('equal')
plt.title('Toss Decision',fontname='Times New Roman',fontsize='42')
plt.show()
```

### Almost 59.3% times teams choose field first

## Toss Decison in each Season

### In many of the seasons most of the times teams after winning toss choose field first

## Win Percentage Field First

```python
total_wins = (cric_data.win_by_wickets>0).sum()
total_loss = (cric_data.win_by_wickets==0).sum()
total = total_wins + total_loss
percent = [(total_wins/total)*100,(total_loss/total)*100]
plt.figure(figsize=(20,8))
plt.pie(percent,
        labels=('Wins','Loss'),
        autopct='%.1f%%',
        colors=('DarkOliveGreen','Yellow'),
        startangle=90,
        shadow=True,
        textprops={'fontsize':30,'fontname':'Times New Roman'})

plt.axis('equal')
plt.title("Win Percentage Field First",fontname='Times New Roman',fontsize=42)
plt.show()
```

### Almost 53.3 times teams choose field first and won the match 

## Won Matches in each season by choosing Field First 

```python
cric_data["field_first"] = "win"
cric_data["field_first"][cric_data['win_by_wickets']==0] = "loss"
plt.figure(figsize=(15,5))
sns.countplot(x='season',hue='field_first',data=cric_data)
plt.xticks(fontsize=12,fontname='Times New Roman')
plt.yticks(fontsize=12,fontname='Times New Roman')
plt.xlabel('Season',fontname='Times New Roman',fontsize=22)
plt.ylabel('Count',fontname='Times New Roman',fontsize=22)
plt.title('Win Matches After Fielding First ',fontname='Times New Roman',fontsize=42)
plt.show()
```

### No. of wins by 'Fileding First' is high compare to 'Batting First'

## Toss Winner is Match Winner

```python
cric_data.toss_winner==cric_data.winner
```

```python
cric_data['toss_comparison']=cric_data.toss_winner==cric_data.winner
cric_data['toss_comparison'].value_counts().plot(kind='pie',
                                                 labels=("YES","NO"),
                                                 figsize=(20,8),
                                                 wedgeprops={'width':0.6,'alpha':0.6},
                                                 autopct='%.1f%%',
                                                 pctdistance=0.75,
                                                 colors=('DarkOliveGreen','Yellow'),
                                                 startangle=90,
                                                 shadow=True,
                                                 textprops={'fontsize':30,'fontname':'Times New Roman'})
plt.axis('equal')
plt.title('Toss Winner Is Match Winner',fontname='Times New Roman',fontsize=42,y=1.1)
plt.show()
```

### Almost 51.3% times toss winner is match winner which is quite high than loss percentage

## Match Result

```python
result_data = cric_data.result.value_counts()
result_data
```

```python
result_data.plot(kind='pie',
                 figsize=(20,9),
                 autopct='%.1f%%',
                 pctdistance=.5,
                 colors=('yellowgreen','Yellow','red'),
                 startangle=None,
                 shadow=True,
                 textprops={'fontsize':30,'fontname':'Times New Roman'},
                 labels=None,
                 explode=(0,.6,.3))
plt.legend(labels=result_data.index,loc='upper right',fontsize=20)
plt.title('Match Result',fontname='Times New Roman',fontsize=50,y=1.1,loc='center')
plt.axis('equal')
plt.show()
```

### Only few of the matches are tie(1%) and comes with no result(0.4%)

## IPL Champions

```python
champ = cric_data.drop_duplicates(subset=['season'],keep='last')[['winner']].reset_index(drop=True)
champ_data = champ.value_counts()
champ_data
```
